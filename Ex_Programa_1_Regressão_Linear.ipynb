```python
#Nome: Arthur Takashi Kanno 11806872
#Nome: Pedro Ian de Almeida Soares 12554327

import pandas as pd
import numpy as np

cars_data = pd.read_csv("vehicles_cleaned_train.csv")
```

```python
cars_data.head()
```

```python
def Conversao(data):
    data_dummies = pd.get_dummies(data)
    data_numpy = data_dummies.to_numpy()

    d = data_numpy[:, 0].reshape(-1, 1)

    x0 = data_numpy[:, 1:]
    x1 = np.hstack((np.ones((x0.shape[0], 1)), x0))
    aux = np.zeros(x1.shape)
    x1 = x1 + aux
    x1[:, 1] = x1[:, 1] - 2001
    X = x1

    X = np.float64(X)
    d = np.float64(d)

    return X, d
```

```python
X, d = Conversao(cars_data)

R = X.T @ X
p = X.T @ d

wo = np.linalg.solve(R, p)
```

```python
cars_test = pd.read_csv("cars_teste.csv")
cars_test.head()
```

```python
X_test, d_test = Conversao(cars_test)

e = d_test - X_test @ wo
erro_quadrado_medio = (sum(e**2))/e.shape[0]
print(erro_quadrado_medio)
```

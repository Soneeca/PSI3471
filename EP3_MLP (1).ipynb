```python
#sigmoid function
def sigmoid(X):
   return 1/(1+np.exp(-X))
```

```python
def derivada_sigmoid(X):
    return X * (1 - X)
```

```python
def redeMLP(x, d, eta, Nt, Nb, Ne, n):

    # número de mini-batches por época
    Nmb = int(np.floor(Nt / Nb))

    # inicialização dos pesos
    Wn_1  = np.random.normal(0, 1, (n, 2))
    W1_2 = np.random.normal(0, (1 / n), (n + 1, 1))
    Yn_1 = np.zeros((n, Nb))
    dphin_1 = np.zeros((n, Nb))

    # passo de adaptação dividido pelo tamanho do mini-batch
    eta = eta / Nb

    Xd = np.hstack((np.ones((Nt, 1)), x, d))

    # for das épocas
    for k in range(Ne):
        np.random.shuffle(Xd)
        X = Xd[:, 0 : 2]
        d = Xd[:, [2]]

        # for dos mini-batches
        for l in range(Nmb):
            dmb = d[l * Nb : (l + 1) * Nb].reshape(-1, 1)
            X1mb = X[l * Nb : (l + 1) * Nb, :]

            #Neurônios da camada oculta
            for i in range(n):
                vn_1 = X1mb @ (Wn_1[i ,:]).reshape(-1, 1)
                Yn_1[i, :] = sigmoid(vn_1).reshape(1, -1)
                dphin_1[i, :] = (derivada_sigmoid((Yn_1[i, :]).reshape(-1, 1))).reshape(1, -1)

            # Neurônio de saída
            X2mb = np.vstack((np.ones((1, Nb)), Yn_1))
            X2mb = X2mb.T
            v1_2 = X2mb @ W1_2
            y1_2 = sigmoid(v1_2)
            dphi1_2 = derivada_sigmoid(y1_2)

            # erro da última camada
            e1mb_2 = dmb - y1_2

            # cálculo dos gradientes locais
            delta1_2 = e1mb_2 * dphi1_2
            W1_2 =  W1_2 + (eta * 2 *(X2mb.T @ delta1_2)).reshape(-1, 1)
            for t in range(n):
                deltan_1 = (dphin_1[i, :].reshape(-1, 1)) * delta1_2 * W1_2[t + 1, 0] #tem que ver o shape do W1_2 mas acho que esta certo
                Wn_1[i, :] = Wn_1[i, :] + (eta* 2* (X1mb.T @ deltan_1)).reshape(1, -1)
    return Wn_1, W1_2
```
